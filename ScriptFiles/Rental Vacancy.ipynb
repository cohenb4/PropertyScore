```python
import pandas as pd
import bs4 as BeatutifulSoup
import requests
```

```python
url = 'https://ipropertymanagement.com/research/rental-vacancy-rate#national-rental-vacancy-rate'
```

```python
vacancy = pd.read_html(url) # Returns list of all tables on page
vacancy = vacancy[0]
```

```python
df = pd.DataFrame(data = vacancy, columns = ['State', 'Vacancy Rate'])
df2 = pd.DataFrame(data = vacancy, columns = ['State.1', 'Vacancy Rate.1'])
df2 = df2.rename(columns = {'State.1': 'State', 'Vacancy Rate.1': 'Vacancy Rate'})
```

```python
vacancy = pd.concat([df, df2])
vacancy = vacancy.sort_values('State')
vacancy = vacancy.reset_index()
vacancy = vacancy.drop(columns = ['index'])
vacancy
```
